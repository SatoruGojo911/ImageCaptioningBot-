# Text Preprocessing

```python
def read_text(path):
  with open(path,'r') as f:
    captions = f.read()
  return captions

def clean_text(sentence):
  # Cleaning text removing non text characters from the word using re
  sentence = sentence.lower()
  sentence = re.sub("[^a-z]+"," ",sentence)
  sentence = sentence.split()

  sentence = [s for s in sentence if len(s)>1]
  sentence = " ".join(sentence)
  return sentence
```

```python
descriptions = {}
for caption in captions:
  id,cap = caption.split('\t')
  id = id[:-2]
  if descriptions.get(id) is None:
    descriptions[id] = []
  descriptions[id].append(cap)
```

# Image Preprocessing

```python
# Importing our ResNet model Backbone
model = tf.keras.applications.ResNet50(weights = "imagenet", input_shape = (224,224,3))
model.summary()
```

```python
# Using only the Fully Convolutional Layers
model_req = tf.keras.Model(model.input,model.layers[-2].output)
```

```python
def pre_processing(img):
  # Preprocessing images according to the resnet model
  img = image.load_img(img,target_size = (224,224,3))
  img = image.img_to_array(img)
  img = np.expand_dims(img,axis = 0)
  #normalizations
  img = preprocess_input(img)
  return img
```

```python
ori_img_path = "Flickr_Data/Flickr_Data/Images/"
```

```python
def encode_image(img):
  # This function gets feature vecture model
  img = pre_processing(img)
  feature_vec = model_req.predict(img)
  feature_vec = feature_vec.reshape((-1,))
  return feature_vec
```

```python
word_to_idx = {}
idx_to_word = {}
for i,word in enumerate(total_words):
  word_to_idx[word] = i+1
  idx_to_word[i+1] = word
print(len(word_to_idx))
word_to_idx['startseq'] = 1846
word_to_idx['endseq'] = 1847
# Adding Start seq and End seq to the vocabulary
idx_to_word[1846] = 'startseq'
idx_to_word[1847] = 'endseq'
vocab_size = len(word_to_idx)+1
```

# Data Generator

```python
##CUSTOM DATA GENERATOR
from tensorflow.keras.utils import pad_sequences
def data_generator(train_desc,train_encoded,word_to_idx,batch_size,maxlength = 35):
  x1,x2,y = [],[],[]
  n = 0
  while True:
    for key,caps in train_desc.items():
      n+=1
      photo = train_encoded[key]
      for cap in caps:
        seq = [word_to_idx[word] for word in cap.split() if word in word_to_idx]
        for i in range(1,len(seq)):
          xi = seq[0:i]
          yi = seq[i]
          xi = pad_sequences([xi],maxlen = maxlength, value = 0, padding = 'post')[0]#0 is the padding value
          yi = to_categorical([yi],num_classes = vocab_size)[0]#one hot vector
          x1.append(photo)
          x2.append(xi)
          y.append(yi)
      if n == batch_size:
        yield ([np.array(x1)],[np.array(x2)]),[np.array(y)]
        x1,x2,y = [],[],[]
        n = 0
```

# Embedding Glove File

```python
glove_file = open("/content/glove.6B.50d1.txt?rlkey=rxrdka5i47xxfnszawwuub2r8")
```

```python
embedding_idx = {}

for line in glove_file:
  values = line.split()
  word = values[0]
  word_embeddings = np.array(values[1:],dtype = 'float')
  embedding_idx[word] = word_embeddings
```

```python
glove_file.close()
```

```python
# Function to get the glove embedding matrix
def get_embedding_matrix():
  embed_dim = 50
  matrix = np.zeros((vocab_size,embed_dim))
  for word,idx in word_to_idx.items():
    embed_vec = embedding_idx.get(word)
    if embed_vec is not None:
      matrix[idx] = embed_vec
  return matrix
```

```python
embed_mat = get_embedding_matrix()
```

#Model Architecture

```python
# Image Model
input_img_features = Input(shape= (2048,))
inp1 = Dropout(0.3)(input_img_features)
inp2 = Dense(256,activation = 'relu')(inp1)
```

```python
# Text model
inp_cap = Input(shape = (35,))
inp_c1 = Embedding(input_dim = vocab_size, output_dim = 50, mask_zero = True)(inp_cap)
inp_c2 = Dropout(0.3)(inp_c1)
inp_c3 = LSTM(256)(inp_c2)
```

```python
from keras.layers import Add
# Combining our image and text models
decoder1 = Add()([inp2, inp_c3])
decoder2 = Dense(256,activation = 'relu')(decoder1)
outputs = Dense(vocab_size,activation = 'softmax')(decoder2)
```

```python
model = Model(inputs = [input_img_features,inp_cap],outputs = outputs)
```

```python
model.summary()
```

```python
model.layers[2].set_weights([embed_mat])
model.layers[2].trainable = False
```

```python
model.compile(loss = 'categorical_crossentropy',optimizer = 'adam')
```

```python
def train():
  # Training the model using model.fit method
  for i in range(10):
    generator = data_generator(train_desc,train_encoded,word_to_idx,batch_size,maxlength = 35)
    model.fit(generator,epochs = 1,steps_per_epoch = steps_per_epoch,verbose = 1 )
    # Saving the model after every epoch
    model.save('/content/Drive/MyDrive/Image_Cap_Models/model_ver1_'+str(i)+'.h5')
```

#Predictions

```python
### PREDICTIONS ###
def predict(photo):
  # The sentence starts with start seq
  in_text = "startseq"
  for i in range(35):
    # Encode the sentence
    sequence = [word_to_idx[w] for w in in_text.split() if w in word_to_idx]
    # Pad the sentences
    sequence = pad_sequences([sequence],maxlen = 35, padding = 'post')
    # Predict what should be the next words
    y_pred = model.predict([photo,sequence])
    # Get the maximum probab predicted word
    y_pred = y_pred.argmax()
    # Convert back to word
    word = idx_to_word[y_pred]
    # Stop if word is endseq
    in_text += (" "+ word)
    if (word == "endseq"):
      break
  # Remove startseq and endseq
  final_cap = in_text.split()
  caption_ans = final_cap[1:-1]
  caption_ans = " ".join(caption_ans)
  return caption_ans
```
